```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Shanghai_HMT_2010.csv')
```

```python
print("data.shape = ", data.shape)
data.head(100)
```

## Удаление элементов типа NaN для очистки данных

```python
data = data.dropna()
```

```python
print("data.shape = ", data.shape)
data.head(10)
```

```python
data = pd.get_dummies(data, drop_first=True)
data = (data - data.mean()) / data.std()
```

```python
print("data.shape = ", data.shape)
data.head(10)
```

## Cоздание тепловой матрицы скоррелированности признаков

```python
list_of_column_names = data.columns
sns.heatmap(data[list_of_column_names].corr(method="pearson"), annot = True, fmt='.1g', vmin=-1, vmax=1, center= 0,
        cmap = "Greens", square=False,
                linewidth=0.5, cbar_kws={"shrink": .5})
```

#### Видно, что в целом мало значений корелируют между собой

## Разбиваем данные тест и трейн

```python
from sklearn.model_selection import train_test_split
```

#### Берем все данные, кроме графы PRES

```python

X_train, X_test, y_train, y_test = train_test_split(data.drop(['PRES'], axis=1), data['PRES'], test_size=0.4)


from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, explained_variance_score

model = LinearRegression(normalize=True)
model.fit(X_train, y_train)
predictions = model.predict(X_test)

mean_squared_error(predictions, y_test)
```

```python
weights = np.zeros(20)
weights[0] = model.intercept_
weights[1:] = model.coef_[:]
```

```python
model.coef_
```

## Строим график, подтверрждающий правильность весов. График проведем по графе TEMP

```python
f = lambda x: weights[0] + weights[12] * x 

x = np.linspace(-2, 7, 100)
y_1 = f(x)

data.plot(x='PRES', y='TEMP', kind='scatter', color = "b")

plt.plot(x, y_1, label='linreg', color='g')

plt.xlabel('PRES')
plt.ylabel('TEMP')
plt.legend(loc='upper left')
plt.show()
```
